# HR Analytics Project

### Load and Previewing the Data
Dataset - [link text](https://www.kaggle.com/datasets/redpen12/employees-satisfaction-analysis)

```python
# Loading relevant libraries and the data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Employee Attrition.csv')
```

```python
# Previewing the top of the dataset
df.head()
```

```python
# Previewing the bottom of the dataset
df.tail()
```

```python
# Checking on the dataset shape
df.shape
```

```python
# Checking on the datatype
df.info()
```

```python
# Checking for missing values
df.isnull().sum()
```

```python
# Drop rows with missing Emp ID
df = df.dropna(subset=['Emp ID'])

```

```python
# Convert categorical columns to 'category' dtype
df['dept'] = df['dept'].astype('category')
df['salary'] = df['salary'].astype('category')
```

```python
# Verify the changes
df.info()
```

```python
#Checking for Duplicates
df.duplicated().sum()
```

### Data Exploration and Visualization

```python
# Summary of the data
df.describe()
```

### Use the whisker boxplot to determine whether there is an outlier in the data.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Boxplot for satisfaction_level
plt.figure(figsize=(8,6))
sns.boxplot(x=df['satisfaction_level'])
plt.title('Boxplot of Satisfaction Level')
plt.show()
```

```python
# Boxplot for last_evaluation
plt.figure(figsize=(8,6))
sns.boxplot(x=df['last_evaluation'])
plt.title('Boxplot of Last Evaluation')
plt.show()
```

```python
# Boxplot for number_project
plt.figure(figsize=(8,6))
sns.boxplot(x=df['number_project'])
plt.title('Boxplot of number of projects')
plt.xlabel('Number of Projects')
plt.ylabel('Frequency')
plt.show()
```

```python
# Boxplot for average_monthly_hours
plt.figure(figsize=(8,6))
sns.boxplot(x=df['average_montly_hours'])
plt.title('Boxplot of average monthly hours')
plt.xlabel('average monthly hours')
plt.ylabel('Frequency')
plt.show()
```

```python
# Boxplot for time_spent_company
plt.figure(figsize=(8,6))
sns.boxplot(x=df['time_spend_company'])
plt.title('Boxplot of time spend company')
plt.xlabel('time spend company')
plt.ylabel('Frequency')
plt.show()
```

### Use the appropriate diagram to demonstrate the categorical variable in the data.

```python
# Countplot for salary distribution
plt.figure(figsize=(8,6))
order = df['salary'].value_counts().index
sns.countplot(y='salary', data=df, order=order)
plt.title('Distribution of Salary')
plt.show()
```

```python
# Countplot for department
plt.figure(figsize=(8,6))
order = df['dept'].value_counts().index
sns.countplot(y='dept', data=df, order=order)
plt.title('Distribution of Department')
plt.show()
```

```python
# Countplot for work_accident
plt.figure(figsize=(8,6))
order = df['Work_accident'].value_counts().head().index
sns.countplot(y='Work_accident', data=df, order=order)
plt.title('Distribution of work accident')
plt.show()
```

### Determine the relationship among the variables.

```python
# Correlation heatmap
plt.figure(figsize=(10,8))
# Select only numerical features for correlation
numerical_df = df.select_dtypes(include=['number'])
correlation = numerical_df.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()

# Pairplot for a few columns
sns.pairplot(df[['satisfaction_level', 'last_evaluation', 'number_project', 'average_montly_hours']])
plt.show()
```

### Plot a necessary graph that you think will aid in understanding the data.

```python
# Histogram for satisfaction_level
plt.figure(figsize=(8,6))
sns.histplot(df['satisfaction_level'], kde=True)
plt.title('Distribution of Satisfaction Level')
plt.show()

# Violin plot for satisfaction_level vs salary
plt.figure(figsize=(8,6))
sns.violinplot(x='salary', y='satisfaction_level', data=df)
plt.title('Satisfaction Level by Salary')
plt.show()
```

```python
# Histogram for last_evaluation
plt.figure(figsize=(8,6))
sns.histplot(df['last_evaluation'], kde=True)
plt.title('Distribution of last evaluation')
plt.show()

# Violin plot for last_evaluation vs salary
plt.figure(figsize=(8,6))
sns.violinplot(x='salary', y='last_evaluation', data=df)
plt.title('last evaluation by Salary')
plt.show()
```

```python
# Histogram for average_monthly_hours
plt.figure(figsize=(8,6))
sns.histplot(df['average_montly_hours'], kde=True)
plt.title('Distribution of average monthly hours')
plt.show()

# Violin plot for average_montly_hours vs salary
plt.figure(figsize=(8,6))
sns.violinplot(x='salary', y='average_montly_hours', data=df)
plt.title('average monthly hours by Salary')
plt.show()
```
